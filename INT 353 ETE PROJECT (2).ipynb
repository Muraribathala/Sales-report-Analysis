# Outliers

```python
import numpy as np

# Function to identify outliers using IQR method
def find_outliers(column):
    q1 = np.percentile(column, 25)
    q3 = np.percentile(column, 75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

# Count outliers for each numeric column
outliers_count = data.select_dtypes(include=['float64']).apply(find_outliers).sum()
print("Count of outliers for each column:")
print(outliers_count)
```

# Hypothesis Testing

```python
# Hypothesis Test for Profit Percentage
from scipy.stats import ttest_1samp

# Null Hypothesis: The average profit percentage is 0.
# Alternative Hypothesis: The average profit percentage is different from 0.
t_stat, p_value = ttest_1samp(data['Profit persentage'], 0)

# Check the p-value
print(f'T-Statistic: {t_stat}, P-Value: {p_value}')

# Interpretation: If p-value < 0.05, reject the null hypothesis.
```

Explanation: This code conducts a one-sample t-test to assess whether the average profit percentage significantly differs from 0. The null hypothesis assumes no difference, and the p-value is checked to determine whether to reject the null hypothesis.

```python
# Hypothesis Test for Sales Quantity Increase
from scipy.stats import ttest_rel

# Check the actual column names in the dataset
sales_columns = [col for col in data.columns if 'Sales Quantity-day' in col]

if len(sales_columns) >= 2:
    # Null Hypothesis: There is no significant increase in sales quantity.
    # Alternative Hypothesis: There is a significant increase in sales quantity.
    t_stat, p_value = ttest_rel(data[sales_columns[0]], data[sales_columns[-1]])

    # Check the p-value
    print(f'T-Statistic: {t_stat}, P-Value: {p_value}')

    # Interpretation: If p-value < 0.05, reject the null hypothesis.
else:
    print("Insufficient columns for hypothesis testing.")
```

Explanation: This code performs a paired t-test to determine whether there is a significant increase in sales quantity between the first and seventh day. The null hypothesis assumes no increase, and the p-value is examined for significance.

```python
# Hypothesis Test for Supplier Profitability
from scipy.stats import f_oneway

# Null Hypothesis: Profitability is the same across different suppliers.
# Alternative Hypothesis: Profitability differs among suppliers.
suppliers = data['Supplier'].unique()
samples = [data[data['Supplier'] == supplier]['Profit persentage'] for supplier in suppliers]
f_stat, p_value = f_oneway(*samples)

# Check the p-value
print(f'F-Statistic: {f_stat}, P-Value: {p_value}')

# Interpretation: If p-value < 0.05, reject the null hypothesis.
```

Explanation: This code conducts an ANOVA test to examine whether the profitability differs among different suppliers. The null hypothesis assumes equal profitability, and the p-value is examined for significance.
